```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import coffee data
coffee = pd.read_csv("../data/crop/coffee_output.csv")

# import weather data
weather = pd.read_csv("../data/weather/weather_data.csv")
```

```python
# TODO: Merge these two dataframes together

merged = pd.merge(weather, coffee, on=['year', 'subdivision'])
```

```python
# TODO: View the first 5 rows of your data

merged.head()
```

```python
# TODO: Check for null values 

merged.isnull().sum()
```

```python
# TODO: Drop rows that contain null values

merged.dropna()
```

```python
# TODO: BEGIN YOUR EDA!

# Evaluate relationship between temp_avg and 60kg bags (scatter plot)

sns.scatterplot(merged, x = "temp_avg", y = "60kgs_bag", color = "darkgreen")


plt.xlabel('Average Temperature in Celsius')
plt.ylabel('Amount of Harvested 60 kg bags(millions)')
plt.title('Average Temperature vs Harvested 60kg bags in Brazil 2006-2023')

plt.show()

#calculate pearson's correlation

pearsons_correlation = merged[['temp_avg', '60kgs_bag']].corr()
print(pearsons_correlation)
```

```python
# Evaluate relationship between temp_avg and amount of bearing trees (scatter plot)

sns.scatterplot(merged, x = "temp_avg", y = "bearing_trees", color = "darkgreen")


plt.xlabel('Average Temperature in Celsius')
plt.ylabel('Amount of Bearing Trees (millions)')
plt.title('Average Temperature vs Amount of Bearing Trees in Brazil 2006-2023')

plt.show()

#calculate pearson's correlation

pearsons_correlation = merged[['temp_avg', 'bearing_trees']].corr()
print(pearsons_correlation)
```
