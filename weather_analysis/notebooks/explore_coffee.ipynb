```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
coffee = pd.read_csv("../data/crop/coffee_output.csv")
```

```python
# TODO: Select only Minas Gerais

mg_coffee = coffee[coffee["subdivision"] == "Minas Gerais"]
mg_coffee.head()
```

```python
# Evaluate the number of bearing trees over time (line plot)

sns.lineplot(mg_coffee, x = "year", y = "bearing_trees", color = "saddlebrown")

plt.xlabel('Year')
plt.ylabel('Amount of Bearing Coffee Trees (million trees)')
plt.title('Amount of Bearing Coffee Trees in Minas Gerais 2002-2023')

plt.show()
```

```python
#histogram 

mg_coffee['bearing_trees'].hist(bins=10, color='saddlebrown')


plt.title('Distribution of Bearing Tres')
plt.xlabel('Bearing Trees')
plt.ylabel('Frequency')

plt.show()
```

```python
 # For my own fun: number of nonbearing trees over time?

sns.lineplot(mg_coffee, x = "year", y = "nonbearing_trees", color = "saddlebrown")

plt.xlabel('Year')
plt.ylabel('Amount of Non-bearing Coffee Trees (million trees)')
plt.title('Amount of Non-bearing Coffee Trees in Minas Gerais 2002-2023')

plt.show()
```

```python
# Evaluate the bearing hectares over time (line plot)

sns.lineplot(mg_coffee, x = "year", y = "bearing_hectares", color = "saddlebrown")

plt.xlabel('Year')
plt.ylabel('Hectares of Bearing Trees ')
plt.title('Amount of Bearing Hectares in Minas Gerais 2002-2023')

plt.show()
```

```python
mg_coffee['bearing_hectares'].hist(bins=10, color='saddlebrown')


plt.title('Distribution of Bearing Hectares')
plt.xlabel('Bearing Hectares')
plt.ylabel('Frequency')

plt.show()
```

```python
# For my own fun: number of bearing trees per hectare over time?

sns.lineplot(mg_coffee, x = "year", y = "bearing_trees_per_hectare", color = "saddlebrown")

plt.xlabel('Year')
plt.ylabel('Amount of Bearing Trees per Hectare')
plt.title('Amount of Bearing Trees per Hectare in Minas Gerais 2002-2023')

plt.show()
```

```python
# Evaluate count of million 60 kgs bags over time (line plot)

sns.lineplot(mg_coffee, x = "year", y = "60kgs_bag", color = "saddlebrown")

plt.xlabel('Year')
plt.ylabel('Amount of Harvested 60 kg bags(millions)')
plt.title('Amount of Harvested 60kg Coffee Bags in Minas Gerais 2002-2023')

plt.show()
```

```python
mg_coffee['60kgs_bag'].hist(bins=10, color='saddlebrown')


plt.title('Distribution of 60kg bags')
plt.xlabel('60kg bags')
plt.ylabel('Frequency')

plt.show()
```

```python
# Evaluate relationship between bearing trees and 60kg bags (scatter plot)

sns.scatterplot(mg_coffee, x = "bearing_trees", y = "60kgs_bag", color = "saddlebrown")


plt.xlabel('Amount of Bearing Coffee Trees (millions)')
plt.ylabel('Amount of Harvested 60 kg bags(millions')
plt.title('Bearing Coffee Trees vs Harvested 60kg bags in Minas Gerais 2002-2023')

plt.show()

#calculate pearson's correlation

pearsons_correlation = mg_coffee[['bearing_trees', '60kgs_bag']].corr()
print(pearsons_correlation)
```

```python
# Evaluate relationship between bearing hecatres and 60 kg bad (scatter)

sns.scatterplot(mg_coffee, x = "bearing_hectares", y = "60kgs_bag", color = "saddlebrown")


plt.xlabel('Amount of Bearing Hectares')
plt.ylabel('Amount of Harvested 60 kg bags(millions')
plt.title('Bearing Hectares vs Harvested 60kg bags in Minas Gerais 2002-2023')

plt.show()

#calculate pearson's correlation

pearsons_correlation = mg_coffee[['bearing_hectares', '60kgs_bag']].corr()
print(pearsons_correlation)
```
