```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
weather1 = pd.read_csv("../data/weather/weather_data1.csv")
weather2 = pd.read_csv("../data/weather/weather_data2.csv")
```

```python
# TODO: concatenate your data files together and save to a new csv file ("../data/weather/weather_data.csv")
weather = pd.concat([weather1, weather2])

weather.to_csv("../data/weather/weather_data.csv", index=False)
```

```python
# TODO: BEGIN YOUR EDA!
# Evaluating how average temperatures change over time for Minas Gerais (line plot)

sns.lineplot(weather, x = "year", y = "temp_avg", color = "royalblue")

plt.xlabel('Year')
plt.ylabel('Average Temperature in Minas Gerais in Celsius')
plt.title('Average Temperature in Minas Gerais 2006-2023')

plt.show()
```

```python
# Evaluating how minimum humidity changes over time for Minas Gerais (line plot)

sns.lineplot(weather, x = "year", y = "hum_min", color = "royalblue")

plt.xlabel('Year')
plt.ylabel('Minimum Humidity %')
plt.title('Minimum Humidity in Minas Gerais 2006-2023')

plt.show()

```

```python
# Evaluating how average wind speed changes over time for Minas Gerais (line plot)

sns.lineplot(weather, x = "year", y = "wind_avg", color = "royalblue")

plt.xlabel('Year')
plt.ylabel('Average Wind Speed (m/s)')
plt.title('Average Wind Speed in Minas Gerais 2006-2023')

plt.show()
```

```python
# Bivariate Analysis: 
# Relationship between max rain and max humidity (scatter plot)

sns.scatterplot(weather, x = "rain_max", y = "hum_max", color = "royalblue")


plt.xlabel('Average Maximum Amount of Rain (mL)')
plt.ylabel('Maximum Humidity %')
plt.title('Maximum Humidity vs Average Maximum Amount of Rain in Minas Gerais 2006-2023')

plt.show()

#calculate pearson's correlation

pearsons_correlation = weather[['rain_max', 'hum_max']].corr()
print(pearsons_correlation)

# find which year had the highest maximum rain fall
#highest_max_rain_fall = 
```

```python
# Bivariate Analysis: 
# Relationship between average wind speed and average temperature (scatter plot)


sns.scatterplot(weather, x = "wind_avg", y = "temp_avg", color = "royalblue")


plt.xlabel('Average Wind Speed (m/s)')
plt.ylabel('Average Temperature in Minas Gerais in Celsius')
plt.title('Average Wind Speed vs Average Temperature in Minas Gerais 2006-2023')

plt.show()

#calculate pearson's coefficient 

pearsons_correlation = weather[['wind_avg', 'temp_avg']].corr()
print(pearsons_correlation)
```
